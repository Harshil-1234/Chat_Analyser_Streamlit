# Step 1: Preprocessing
Chat is present as a text file. Pehle text ko extract krke pandas ka dataframe bnaenge

```python
import re # re -> regular expressions
import pandas as pd
```

```python
f = open('WhatsApp Chat with Sarthak.txt','r',encoding = "utf-8")
```

```python
data = f.read()
```

```python
print(data) # whole whatsapp chat is stored as a string in variable 'data'
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2},\s\d{1,2}:\d{2}\s(?:AM|PM)\s-\s'
# above is a regular expression to separate content before the names
# site to buld reg exp = regx101
```

```python
messages = re.split(pattern,data)[1:]
messages
```

```python
dates = re.findall(pattern,data)
dates = [string.replace("\u202f", " ") for string in dates]
dates
```

```python
#creating a dataframe with two columns
df = pd.DataFrame({'user_message': messages, 'message_date': dates})
# convert message_date type
df['message_date'] = pd.to_datetime(df['message_date'], format='%m/%d/%y, %I:%M %p - ')

df.rename(columns={'message_date': 'date'}, inplace=True)

df.head()
```

```python
df.shape
```

```python
#separate user name and messages
users = []
messages = []

for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:  # user name
        users.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages

df.drop(columns=['user_message'],inplace = True)

df.head()
```

```python
df['year'] = df['date'].dt.year
```

```python
df.head()
```

```python
df['month'] = df['date'].dt.month_name()
```

```python
df.head()
```

```python
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
```
